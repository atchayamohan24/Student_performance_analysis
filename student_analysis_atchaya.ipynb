```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df=pd.read_csv("student-mat.csv")
df.head()
```

```python
#Explore data
df.shape
```

```python
#Checking missing values
df.isnull().sum()
```

```python
#Check duplicates
df.duplicated().sum()
```

```python
#Analysis

#Average final grade
df["G3"].mean()
```

```python
#Analysis

#Student scored above 15
(df["G3"]>15).sum()
```

```python
#Analysis

#Study time vs Grade
df[["studytime", "G3"]].corr()
```

```python
#Analysis

#Gender performance
df.groupby("sex")["G3"].mean()
```

```python
#Visualization

#histogram
plt.hist(df["G3"],bins=10)
plt.show()
```

```python
#Visualization

#Scatter plot
plt.scatter(df["studytime"],df["G3"])
plt.show()
```

```python
#Visualization

#Bar chart
df.groupby("sex")["G3"].mean().plot(kind="bar")
plt.show()
```
